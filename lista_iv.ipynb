Faça um Programa que leia 20 números inteiros e armazene-os numa lista. Armazene os números pares na lista PAR e os números IMPARES na lista impar. Imprima as três listas.

```python
todos = []
par = []
impar = []

for i in range(20):
  numero = int(input('Número: '))
  todos.append(numero)

  if numero % 2 == 0:
    par.append(numero)

  else:
    impar.append(numero)


print('Todos os números: ',todos)
print('Par', par)
print('Impar', impar)
```

5. Faça um Programa que peça as quatro notas de 10 alunos, calcule e armazene numa lista a média de cada aluno, imprima o número de alunos com média maior ou igual a 7.0.

```python
notas = []
media_aprovado = 0

for i in range(10):
  quatro_notas = []

  for j in range(4):
    nota = float(input('Digite as notas: '))
    quatro_notas.append(nota)

  media = sum(quatro_notas)/4
  notas.append(media)
  print(media)

  media_aprovado = 0
  if media >= 7:
   media_aprovado += 1

print('Média: ', notas)
print('Alunos com média maior ou igual a 7.0: ', media_aprovado)



```

6. Faça um Programa que leia e armazene 50 números inteiros, mostre a soma, a multiplicação e os números.

```python
numeros = []

for i in range(50):
  numero = int(input('Digite os números: '))
  numeros.append(numero)

  soma = sum(numeros)
  mult = 1
  for num in numeros:
    mult *= num

print('soma: ', soma)
print('multiplicação: ', mult)
print(numeros)
```

7. Faça um Programa que peça a idade e a altura de 10 pessoas, armazene cada informação na sua respectiva lista. Imprima a idade da pessoa que possui maior altura.

```python
idades = []
alturas = []

for i in range(10):
  idade = int(input('idade: '))
  idades.append(idade)

  altura = float(input('altura: '))
  alturas.append(altura)

maior_altura = max(alturas)
indice_maior_altura = alturas.index(maior_altura)
maior_idade = idades[indice_maior_altura]

print('idades: ',idades)
print('alturas: ',alturas)
print('maior idade: ',maior_idade)
```

8. Faça um Programa que leia uma lista A com 10 números inteiros, calcule e mostre a soma dos quadrados dos elementos do vetor.

```python
a = []

for i in range(10):
  numero = int(input('Digite um número: '))
  a.append(numero)

  soma_quadrados = 0
  for j in a:
    soma_quadrados += j ** 2

print('soma: ', soma_quadrados)
```

11. Foram anotadas as idades e alturas de 30 alunos. Faça um Programa que determine quantos alunos com mais de 13 anos possuem altura inferior à média de altura desses alunos.

```python
idades = []
alturas = []
contador = 0

for i in range(30):
  idade = int(input("Digite sua idade: "))
  idades.append(idade)

for i in range(30):
  altura = float(input("Digite sua altura: "))
  alturas.append(altura)

media_altura = sum(alturas)/3

for i in range(30):
    if idades[i] > 13 and alturas[i] < media_altura:
        contador += 1

print('média de alturas: ', media_altura)
print('alunos que tem mais de 13 anos e altura inferior a média: ', contador)
```

12. Escreva um algoritmo que permita a leitura dos nomes de 10 pessoas e armazene os nomes lidos em uma lista. Após isto, o algoritmo deve permitir a leitura de mais 1 nome qualquer de pessoa e depois escrever a mensagem ACHEI, se o nome estiver entre os 10 nomes lidos anteriormente (guardados na lista), ou NÃO ACHEI caso contrário.

```python
nomes = []

for i in range(10):
  nome = input('Digite seu nome: ')
  nomes.append(nome)

novo_nome = input('digite mais um nome: ')

if novo_nome in nomes:
  print('ACHEI!')

else:
  print('NÃO ACHEI! ')

print(nomes)
```

13. Faça um programa que receba a temperatura média de cada mês do ano e armazene-as em uma lista. Após isto, calcule a média anual das temperaturas e mostre todas as temperaturas acima da média anual, e em que mês elas ocorreram (mostrar o mês por extenso: 1 – Janeiro, 2 – Fevereiro, . . . ).

```python
meses = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho", "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"]

temperaturas = []

for i in range(12):
  cada_mes = float(input(f'digite a temperatura média de {meses[i]}: ' ))
  temperaturas.append(cada_mes)

media_anual = sum(temperaturas)/len(temperaturas)
print('média anual', media_anual)
for i in range(12):
  if temperaturas[i] > media_anual:
    print(f"{meses[i]}: {temperaturas[i]}")
```
